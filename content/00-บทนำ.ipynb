เนื่องจากว่า อัตราเร็วของวัตถุที่เคลื่อนที่ในแนวดิ่งเปลี่ยนร็วมาก  การทดลองทำโดยใช้พื้นเอียงที่เอียงเพื่อให้การเพิ่มของอัตราเร็วไม่มากเกินไปที่จะวัดปริมาณที่สนใจได้

ในการวัดจำเป็นต้อง มีหน่วยสำหรับวัด ในตัวอย่างนี้หน่วยวัดความยาว และ เวลา ไม่ใช่หน่วยสากลที่นิยมใช้กัน แต่ เป็นหน่วยที่สะดวกต่อการทดลอง

## ผลทดลอง และ การวิเคราะห์

ในการวิเคราะห์ จะทำโดยใช้ code ภาษา python ซึ่งเป็นภาษาที่นิยมใช้อย่างแพร่หลาย และ มีเครื่องมือที่จำเป็นสำหรับการวิเคราะห์ที่เราต้องทำ สิ่งแรกที่ต้องทำเพื่อใช้เครื่องเหล่านี้คือ ต้องนำมันขึ้นไปในหน่วยความจำ โดยใช้คำสั่งที่ตัวภาษาเตรียมไว้ให้ คือ คำสั่ง pip import สี่บรรทัดแรกของ cell ถัดไป คือการนำกลุ่มของเครื่องมือที่เรียกในภาษานี้ว่า "โมดูล" เข้ามาใช้งาน

- โมดูลหลัก matpltlib กลุ่มของเครื่่องมือที่ใช้ทำกราฟ
- โมดูย่อย pyplot ในโมดูลหลัก matpltlib
- โมดูลหลัก numpy กล่มของเครื่องมือที่ใช้ทำการคำนวณ

จะสังเกตเห็นว่าคำสั่งใน import มี คำว่า as เป็นคำที่ใช้กำหนดชื่อย่อของโมดูล เพื่อความสะสดวกในการใช้งานเครื่องมือใน โมดูล เหล่านั้น (ไม่ต้องอ้างอิงโดยใช้ชื่อเต็ม)
การใช้เครื่องมือใน โมดูล ทำโดยใช้ชื่อ ย่อ นี้ ตามด้วย จุด "." และ ชื่อของเครื่องมือ 

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=1)
np.set_printoptions(floatmode="fixed")
mpl.font_manager.fontManager.addfont("LeelawUI.ttf")
```

```python
# กำหนดระยะทางการเคลื่อนที่  4 ระยะที่แตกต่างกัน คือ 1, 2, 3, 4 L เก็บไว้ที่ตัวแปร s_exp
s = np.array([1.0, 2.0, 3.0, 4.0]) 

# ผลการวัดเวลาซึ่งวัดโดยมวลน้ำที่ไหลระหว่างที่วัดถุเคลื่อนที่ 
# เก็บไว้ที่ตัวแปร "มวลแก้วและน้ำ" ซึ่งนำไปคำนวณเป็นเวลา ในหน่วย T และเก็บไว้ที่ t_exp

มวลแก้วเปล่า = 29.3 # กรัม
มวลแก้วและน้ำ = np.array([34.2, 53.7, 62.7, 77.3]) # กรัม

มวลน้ำ = มวลแก้วและน้ํา - มวลแก้วเปล่า
# แปลงเวลาในหน่วย มวลน้ำ เป็นหน่วย รอบของลูกตุ้ม
อัตราการไหลฯ = 10 # กรัม ต่อ รอบ  
t = มวลน้ำ/อัตราการไหลฯ

print(s, "s (L)");print(t, "t (T)")
```

### การวิเคราะห์
#### ทฤษฎี 
>ระยะทาง s แปรผันตรง กับ กำลังสองของเวลา t  
$$s \; \alpha \; t^2 \;\; หรือ \;\; s = c t^2 + s_0$$ 
$$ c = ค่าคงที่การแปรผัน \;(หรือ ครึ่งหนึ่งของความเร่ง\; \frac{1}{2} a)\; \;\;, s_0 = ตำแหน่งเริ่มต้น$$

#### วิเคราะโดยอาศัยกราฟของผลการวัด และ กราฟของทฤษฎี

การวิเคราะห์อาศัยความจริงที่ว่า กราฟของความสัมพันธ์ของตัวแปรตามที่แปรผันตรงกับตัวแปรอิสระจะเป็นเส้นตรง 
ตามทฤษฎีที่เราสันนิษฐานไว้ ระยะ $s$ จะแปรผันตรงกับกำลังสองของเวลา $t^2$ ดังนั้นกราฟระหว่างสองตัวแปรนี้ จะเป็นเส้นตรง

```python
# คำนวณกำลังสองของเวลา ที่ได้จากการทดลอง (เก็บไว้ในตัวแปร t_exp) เก็บ
# ผลการคำนวณไว้ที่ tsqr_exp
t_sq = t**2

# ค่าคงที่การแปรผัน หรือ ความชันของเส้นตรง
c = (s[2] - s[1])/(t_sq[2] - t_sq[1])

# หาค่าเริ่มต้น s_0
s_0 = s[1] - c*t_sq[1]

# กำหนดช่วงเวลาสำหรับการคำนวณ คือ เริ่มจาก 0 ไปสิ้นสุดที่ 5  โดยเพิ่มที่ละ 
# 0.05 (5/100) เก็บช่วงเวลา นี้ไว้ใน t_th
t_th = np.linspace(0,5,100)


# คำนวณระยะ s จากทฤษฎี เก็บไว้ที่ s_th
s_th = c*(t_th**2) + s_0
# ทำกราฟ ระหว่าง t^2 และ s
plt.scatter(t_sq, s) # ทำกราฟผลการทดลอง 
plt.plot(t_th**2, s_th) # ทำกราฟทฤษฎี

# เพิ่มข้อมูลให้แก่กราฟ
plt.title("กราฟที่ 1\nความสัมพันธ์ระหว่าง t^2 และ s ")
plt.legend(["ผลการทดลอง", f"ทฤษฎี (ความชัน={c:.1f} จุดตัดแกน y ={s_0:.1f})"])
plt.xlabel("t^2 (T^2)")
plt.ylabel("s (L)")
plt.axis([0,(t[3] + 1)**2,0,s[3] + 1]);
```

## สรุป และ ใช้ทฤษฎีทำนาย ปรากฏการณ์ที่เกี่ยวของ
จากกราฟที่ 1 สามารถสรุปได้ว่า ทฤษฎีนี้ใกล้เคียงกับความเป็นจริง กล่าวคือ ระยะทาง $s$ มีความสัมพันธ์กับเวลา $t$ ในแบบกำลังสอง

$$s \; \alpha \; t^2$$ 

จากข้อสรุปนี้ เราสามารถทำนายได้ว่าหลังจากปล่อยวัตถุลงจากพื้นเอียงแล้วมันอยู่ที่ไหนบนพื้นเอียง เพื่อให้การพิสูจน์การทำนายทำได้ง่าย นิยยามหนึ่งหน่วยความยาว และ หนึ่งหน่วยเวลาใหม่ดัง นี้

>หนึ่งหน่วยความยาว เท่ากับ 1 ใน 16 ของความยาวพื้นเอียง

>หนึ่งหน่วยเวลา เท่ากับ ช่วงเวลาที่วัตถุใช้เคลื่อนที่จากจุดหยุดนิ่ง ออกไปเป็นระยะ 1 หน่วยความยาว

โดยใช้นิยามใหม่นี้ และ ข้อสรุปที่ว่า $s \; \alpha \; t^2 \;\;$  สามารถทำนายได้ว่า

> $$ t = 1 \; s = 1 $$ 
> $$t = 2 \; s = 4 $$ 
> $$ t = 3 \; s = 9 $$ 
> $$ t = 4 \; s = 16  $$
